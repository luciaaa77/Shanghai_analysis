# Radial plot 

```python
# coding=utf-8
import numpy as np
import matplotlib.pyplot as plt

# the number of the facilities is taking logarithsm and done in the excel 
results = [{ "Kindergarten*": 1.04453976,  "parks*":0.35 , "Hospital": 3.9,"primary school*": 0.71,"Top middle school": 0.87,"shopping centre": 4.5,"underground station": 1.29,"High-speed train station": 0.08},
           {"Kindergarten*": 1.2,  "parks*": 0.71, "Hospital": 0,"primary school*": 0.77,"Top middle school": 0.25,"shopping centre": 0,"underground station": 0.15,"High-speed train station": 0}]
data_length = len(results[0])
# The polar coordinates are divided equally according to the length of the data
angles = np.linspace(0, 2*np.pi, data_length, endpoint=False)
labels = [key for key in results[0].keys()]
score = [[v for v in result.values()] for result in results]
# close the radial plot 
score_a = np.concatenate((score[0], [score[0][0]]))
score_b = np.concatenate((score[1], [score[1][0]]))
angles = np.concatenate((angles, [angles[0]]))
labels = np.concatenate((labels, [labels[0]]))
# size 
fig = plt.figure(figsize=(8, 6), dpi=100)
# subplot 
ax = plt.subplot(111, polar=True)
# plotting 
ax.plot(angles, score_a, color='g')
ax.plot(angles, score_b, color='b')
# 
ax.set_thetagrids(angles*160/np.pi, labels)
# 
ax.set_theta_zero_location('N')
# set the range 
ax.set_rlim(0, 5)
# set the angle 
ax.set_rlabel_position(270)
#ax.set_title("Radio plot for average built environment facilities in each clsuter")
plt.legend(["HH cluster", "LL cluster"], loc=4)
plt.savefig('radial.png')
```
